## Data Encoding

```python
# List of variables to map

varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
housing[varlist] = housing[varlist].apply(binary_map)
```

```python
# Check the housing dataframe now

housing.head()
```

```python
# Get the dummy variables for the feature 'furnishingstatus' and store it in a new variable - 'status'
status = pd.get_dummies(housing['furnishingstatus'])
# Check what the dataset 'status' looks like
status.head()
```

```python
# Let's drop the first column from status df using 'drop_first = True'
status = pd.get_dummies(housing['furnishingstatus'], drop_first = True)
```

```python
# Add the results to the original housing dataframe (conacatenate = combining the dataframe)

housing = pd.concat([housing, status], axis = 1)
```

```python
#lets check the results 

housing.head()
```

### Droping the column

```python
# Drop 'furnishingstatus' as we have created the dummies for it

housing.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
#checking the results after droping the "furnishingstatus" column

housing.head()
```

## Splitting the Data into Training and Testing Sets

```python
housing.shape
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(housing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True,cmap="YlOrRd")
plt.show()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

## Model Selection

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Create an instance of Linear Regression
lm = LinearRegression()

# Initialize RFE with a Linear Regression model and the number of features you want to select (e.g., 6)
rfe = RFE(estimator=lm, n_features_to_select=6)

# Fit RFE on your training data
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

## Model Training

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
res = (y_train_price - y_train)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

```python
plt.scatter(y_train,res)
plt.show()
```

```python
# There may be some relation in the error terms.
```

# Model Evaluation

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)
```

```python
# Now let's use our model to make predictions.
```

```python
# Creating X_test_new dataframe by dropping variables from X_test
X_test_rfe = X_test[X_train_rfe.columns]
```

```python
# Making predictions
y_pred = lm.predict(X_test_rfe)
```

```python
#Importing metrix library 

from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)  
```
